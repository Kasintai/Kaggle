```python
import pandas as pd
import numpy as np

# Replace 'your_file.csv' with the actual path to your CSV file
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_train
```

```python
# Create a column of 'Policy Start Date Year'

# Assuming 'Policy Start Date' column exists and is in datetime format
df_train['Policy Start Year'] = pd.to_datetime(df_train['Policy Start Date']).dt.year
df_train['Policy Start Month'] = pd.to_datetime(df_train['Policy Start Date']).dt.month
df_train['Policy Start Day'] = pd.to_datetime(df_train['Policy Start Date']).dt.day

df_test['Policy Start Year'] = pd.to_datetime(df_test['Policy Start Date']).dt.year
df_test['Policy Start Month'] = pd.to_datetime(df_test['Policy Start Date']).dt.month
df_test['Policy Start Day'] = pd.to_datetime(df_test['Policy Start Date']).dt.day
df_test
```

```python
# Drop 'policy start date'

# Drop the 'Policy Start Date' column from both training and testing datasets
df_train = df_train.drop(['id', 'Policy Start Date'], axis=1)
df_test = df_test.drop(['id', 'Policy Start Date'], axis=1)
```

```python
print(df_train.isnull().sum(), '\n')
print(df_train.dtypes, '\n')

print(df_test.isnull().sum(), '\n')
print(df_test.dtypes)
```

```python
import numpy as np

# Add IncomePerDependent
df_train['Income per Dependent'] = df_train['Annual Income'] / df_train['Number of Dependents']

# Handle division by zero and NaN values
df_train['Income per Dependent'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add FinancialStability
df_train['Financial Stability'] = df_train['Credit Score'] * np.log(df_train['Annual Income'])

# Handle negative or zero income values before applying log
df_train['Financial Stability'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add ClaimsRate
df_train['Claims rate'] = df_train['Previous Claims'] / df_train['Insurance Duration']
df_train['Claims rate'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add VehicleAgeRatio
df_train['Vehicle Age Ratio'] = df_train['Vehicle Age'] / df_train['Age']
df_train['Vehicle Age Ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)

# # Replace NaN values with zeros or appropriate values if needed
# df_train.fillna(0, inplace=True)
```

```python
import numpy as np

# Add Income per Dependent
df_test['Income per Dependent'] = df_test['Annual Income'] / df_test['Number of Dependents']

# Handle division by zero and NaN values
df_test['Income per Dependent'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add Financial Stability
df_test['Financial Stability'] = df_test['Credit Score'] * np.log(df_test['Annual Income'])

# Handle negative or zero income values before applying log
df_test['Financial Stability'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add Claims rate
df_test['Claims rate'] = df_test['Previous Claims'] / df_test['Insurance Duration']
df_test['Claims rate'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Add Vehicle Age Ratio
df_test['Vehicle Age Ratio'] = df_test['Vehicle Age'] / df_test['Age']
df_test['Vehicle Age Ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)

# # Replace NaN values with zeros or appropriate values if needed
# df_test.fillna(0, inplace=True)
```

# Model:

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_log_error

# Assuming 'df_train' is your DataFrame and 'Premium Amount' is the target variable
X = df_train.drop('Premium Amount', axis=1)
y = df_train['Premium Amount']

y_log = np.log1p(y)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y_log, test_size=0.2, random_state=42)

# Set XGBoost parameters
xgb_params = {
    'objective': 'reg:squarederror',  # Regression task
    'n_estimators': 100,              # Number of boosting rounds (trees)
    'learning_rate': 0.05,            # Step size shrinkage
    'max_depth': 6,                   # Maximum depth of a tree
    'min_child_weight': 1,            # Minimum sum of instance weight needed in a child node
    'gamma': 0.1,                     # Minimum loss reduction required to make a split
    'subsample': 0.8,                 # Fraction of samples used for each boosting round
    'colsample_bytree': 0.8,          # Fraction of features used for each boosting round
    'lambda': 1.0,                    # L2 regularization term (Ridge regularization)
    'alpha': 0.0,                     # L1 regularization term (Lasso regularization)
    'random_state': 42,               # Seed for reproducibility
    'early_stopping_rounds': 50       # Early stopping criteria
}

# Initialize the XGBoost model with the parameters
model = xgb.XGBRegressor(**xgb_params)

# Train the model with early stopping
model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    #verbose=100,                 # Print updates every 100 rounds
    #early_stopping_rounds=50     # Stop if validation error doesn't improve for 50 rounds
)
```

```python
# Make predictions on the validation set
y_pred_log = model.predict(X_val)
y_pred = np.expm1(y_pred_log)

y_val_cv = np.expm1(y_val)

# Evaluate the model
rmsle = np.sqrt(mean_squared_log_error(y_val_cv, y_pred))
print(f'RMSLE on validation set: {rmsle}')
```

```python
# prompt: feature importance

# Get feature importances from the trained XGBoost model
feature_importances = model.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print or display the feature importances
feature_importance_df
```
